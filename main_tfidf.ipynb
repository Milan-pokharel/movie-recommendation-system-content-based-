```python
#AST Safely converts a string into Python objects (list, dict, tuple).
import ast
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head(3)
```

```python
movies.iloc[0].keywords
```

```python
#for keyword column:
movies['keywords']=movies['keywords'].apply(convert)
```

```python
# taking the top 3 cast members only:
def convert2(obj):
    l=[]
    count=0
    for i in ast.literal_eval(obj):
        if count!=3:
            l.append(i['name'])
        else:
            break
        count+=1
    return l
```

```python
movies['cast']=movies['cast'].apply(convert2)
```

```python
movies.head(2)
```

```python
movies.iloc[1].crew
```

```python
#working on crew column to get the director name:
def get_director(obj) :
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew']=movies['crew'].apply(get_director)
```

```python
movies["overview"]=movies["overview"].apply(lambda x:x.split())
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
#lowering the text:
df['tags']=df['tags'].apply(lambda x:x.lower())
```

```python
#checking for both count vectorizer and tfidf vectorizer,one with better result will be used finally:



#using count vectorizer(bow):
from sklearn.feature_extraction.text import CountVectorizer
bow=CountVectorizer(max_features=7000,stop_words='english')



#using tfidf vectorizer instead of count vectorizer(bow):
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(
    stop_words='english',
    max_features=8000,
    ngram_range=(1,2)   # including bigrams
)




```

```python
#applying stemming to reduce the words to their root form:
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
ps.stem('runnning')
```

```python
#function to perform stemming to make sure different forms of same word are treated equally:
def stem(text):
    y=[]
    for i in  text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df['tags']=df['tags'].apply(stem)
```

```python
df['tags'][2]
```

```python
#creating the vectors now:
vectors_bow = bow.fit_transform(df['tags']).toarray()
vectors_tfidf = tfidf.fit_transform(df['tags']).toarray()
```

```python
vectors_tfidf
```

```python
import numpy as np
np.set_printoptions(threshold=np.inf)
tfidf.get_feature_names_out()
```

```python
#calculating the similarity:
from sklearn.metrics.pairwise import cosine_similarity  

similarity_bow=cosine_similarity(vectors_bow)
similarity_tfidf=cosine_similarity(vectors_tfidf)
```

```python
#recommendation system using bow:
def recommend_bow(movie):
    try:
        movie_index=df[df['title']==movie].index[0]
        distances=similarity_bow[movie_index]
        movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:16]
        
        for i in movies_list:
            print(df.iloc[i[0]].title)
    except IndexError:
        print("Movie not found. Please check spelling.")
                    
print("Movies recommended using Bag of Words:\n")
recommend_bow('The Conjuring')
```

```python

#recommendation system using tfidf:
def recommend_tfidf(movie):
    movie_index=df[df['title']==movie].index[0]
    distances=similarity_tfidf[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:16]
    
    for i in movies_list:
        print(df.iloc[i[0]].title)

print("Movies recommended using TF-IDF:\n")
recommend_tfidf('The Conjuring')
```

```python
#Compare Similarity Statistics:
import numpy as np
print("TF-IDF mean similarity:", np.mean(similarity_tfidf))
print("TF-IDF std:", np.std(similarity_tfidf))

print("\nBOW mean similarity:", np.mean(similarity_bow))
print("BOW std:", np.std(similarity_bow))
```

```python
#looking and analysing all the similarity values of both tfidf and bow, 
#tfidf approach seems to be better so we will use tfidf for our final project and importing its pickle similation scores:

import pickle
pickle.dump(df,open('movies.pkl','wb'))
pickle.dump(df.to_dict(),open('movies_dict.pkl','wb'))
pickle.dump(similarity_tfidf,open('similarity_tfidf.pkl','wb'))
```
